```python
#IMPORT-ant packages
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv


#Entries 
prog_name=[]
prog_overview=[]
prog_outline=[]
prog_duration=[]
prog_tutionUK=[]
prog_tutionUS=[]
careers= []
```

```python
#Stantiating Our Driver

PATH = 'C:\SeleniumDriver\chromedriver.exe'
driver = webdriver.Chrome(PATH)
driver.get('https://www.lboro.ac.uk/study/')
driver.maximize_window()
links_list=[]


#Eating Cookies
try:
    cookies = WebDriverWait(driver, 3).until(
        EC.presence_of_element_located((By.ID, "ccc-dismiss-button"))
    )
    cookies.click()
except:
    driver.quit()

#Capturing Undergraduate Programms links
under_grad= driver.find_element(By.CLASS_NAME, 'button__text')
under_grad.click()

plinks = driver.find_elements(By.CSS_SELECTOR, 'a.list__link')
for link in plinks:
    if str(link.get_attribute('href')).startswith("https://www.lboro.ac.uk/study/undergraduate/courses"): 
        links_list.append(link.get_attribute('href'))

courss_links= (links_list[7:])

#finding elements in every course page 
for course in courss_links:
    driver.get(course)
    prog_name.append(driver.find_element(By.CLASS_NAME,'course-info__heading').text)
    prog_duration.append(driver.find_element(By.CLASS_NAME,'course-info__subheading').text)
    prog_tutionUK.append(driver.find_element(By.CSS_SELECTOR,'div.list__text--uk-eu').text[5:])
    prog_tutionUS.append(driver.find_element(By.CSS_SELECTOR,'div.list__text--international').text[16:])

    time.sleep(2)

    driver.execute_script("window.scrollBy(0,1180)","")

#finding programm overview
    overview=""
    lead = driver.find_element(By.CLASS_NAME, 'lead-paragraph').text
    overview += lead


    o_div = driver.find_element(By.CLASS_NAME,'editor')
    o_ps = o_div.find_elements(By.XPATH,'./p')
    for op in o_ps:
        overview = overview + op.text

    prog_overview.append(overview)


#finding what you'll study

    driver.execute_script("window.scrollBy(0,3820)","")
    study_obj=''
    div = driver.find_element(By.CLASS_NAME,'editor')
    ps = div.find_elements(By.XPATH,'./p')
    for p in ps:
        study_obj = study_obj + p.text

    prog_outline.append(study_obj)


#Extracting Projected career

    driver.execute_script("window.scrollBy(0,12800)","")

    careers.append(driver.find_element(By.CLASS_NAME, 'lead-paragraph').text)


#Creating a CSV file with Extracted info
zipped = zip(prog_name, prog_duration, prog_overview, prog_outline, prog_tutionUK, prog_tutionUS, careers)
with open('Iboro_uni.csv', 'w', newline="") as f:
        writer = csv.writer(f)
        writer.writerow(['Programm Name','Duration', 'Overview', 'Outline','UK_Tution', 'Internation_Tution', 'Projected_career'])
        for name, du, ove, out, uk, inter, career in zipped:
            writer.writerow([name, du, ove, out, uk, inter, career])

```
